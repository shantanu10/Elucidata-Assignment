## Operation 2

```python
#Adding a new column in the parent dataset with the name “Retention Time Roundoff (in mins)
def func(row):
    return round(row['Retention time (min)'])
data["Retention Time Roundoff (in mins)"] = data.apply(func,axis=1)

data.head(5)

#operation 2 ended
```

## operation 3

```python
"""
find the mean of all the metabolites which have same "Retention Time Roundoff"
across all the samples. The resultant of above operation should be a new data-frame where you have to
include the "Retention Time Roundoff" column and all samples.
"""
col_list = list(data.columns)
col_list = [e for e in col_list if e not in {'Accepted Compound ID','Retention time (min)','m/z'}]
col_list
```

```python
new_data = data[col_list]
new_data.columns
```

```python
col_list.remove('Retention Time Roundoff (in mins)')
new_data = new_data.groupby('Retention Time Roundoff (in mins)')[col_list].mean()
new_data.head(7)

#operation 3 ended
```
